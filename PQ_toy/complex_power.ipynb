```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

%matplotlib inline
```

```python
dt = 1/1e4
time = np.arange(0, 1/10, dt)

f = 60 # Hz
w = 2*np.pi*f

# voltage
v = np.cos(w*time)

theta_slider = widgets.FloatSlider(
    value=0.0,
    min=-np.pi/2,
    max=np.pi/2,
    step=0.1,
    description=r'theta',
    continuous_update=True,
)

def plot_v_i(theta):
    #plt.clf()                 # clear previous plot
    # set figure size
    plt.figure(figsize=(10, 6))
    i = np.cos(w * time + theta)
    plt.plot(time, v, label='$v(t)$')
    plt.plot(time, i, label='$i(t)$', linestyle='--')
    plt.plot(time, v*i, label='$p(t)$')
    mean_power = np.mean(v*i)
    plt.axhline(mean_power, color='green', linestyle='--', label='mean power')
    plt.legend(loc='upper right')
    plt.xlabel('time')
    plt.ylabel('amplitude')
    plt.grid(True)
    plt.show()

out = widgets.interactive_output(plot_v_i, {"theta": theta_slider})

display(theta_slider, out)
```
